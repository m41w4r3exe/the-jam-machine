```python
import pandas as pd
```

```python
# load handpicked artists
df = pd.read_csv('../data/artist_genre_handpicked.csv', sep=';')

# reduce size of df to only include picked artists
df.drop("pop thatâ€™s like electro", axis=1, inplace=True)

# create a list of electronic artists
electronic_artists = df[df['electronic'].notna()]['artist']

# create a list of good music artists
good_music_artists = df[df['good music'].notna()]['artist']

# Remove artists from good music list that are also in electronic list
good_music_artists = good_music_artists[~good_music_artists.isin(electronic_artists)]

good_music_artists
```

```python
# load mmd scraped artist genre
df = pd.read_csv('../data/mmd_scraped_artist_genre.csv')

# keep only artists that are in the electronic list
df_electronic = df[df['artist'].isin(electronic_artists)]

# keep only artists that are in the good music list
df_good_music = df[df['artist'].isin(good_music_artists)]

# remove md5 duplicates
df_electronic.drop_duplicates(subset=['md5'], inplace=True)
df_good_music.drop_duplicates(subset=['md5'], inplace=True)

# export both lists
df_electronic.to_csv('../data/electronic_artists.csv', index=False)
df_good_music.to_csv('../data/good_music_artists.csv', index=False)
```

```python
from fuzzywuzzy import fuzz
import itertools

# get all david bowie songs
df_bowie = df_electronic[df_electronic['artist'] == 'David Bowie']

# compute all pairwise combinations of artist_title
bowie_combinations = list(itertools.combinations(df_bowie.title, 2))

# turn combinations into a data frame
df_bowie_combinations = pd.DataFrame(bowie_combinations, columns=['artist_title_1', 'artist_title_2'])

# compute the similarity ratio
df_bowie_combinations['similarity'] = df_bowie_combinations.apply(lambda x: fuzz.partial_ratio(x['artist_title_1'], x['artist_title_2']), axis=1)

df_bowie_combinations

# export df_bowie_combinations
df_bowie_combinations.to_csv('../data/bowie_combinations.csv', index=False)
```

```python
import pandas as pd
# extract all filnames from a folder
import os
import shutil

input_path = '../data/music_picks/good_music_artists'
output_path = '../data/music_picks/good_music_deduped'
reference_path = '../data/music_picks/model_statistics_deduplicated.csv'

# create output folder if it does not already exist
if not os.path.exists(output_path):
    os.makedirs(output_path)

# get all file paths from the folder and subfolders and store them in a list
file_paths = [os.path.join(dp, f) for dp, dn, filenames in os.walk(input_path) for f in filenames]

# create a data frame with the file paths
df = pd.DataFrame(file_paths, columns=['file_path'])

# create column with the filename
df['filename'] = df['file_path'].apply(lambda x: x.split('/')[-1])

# keep only .mid files
df = df[df['filename'].str.contains('.mid')]

# create column with the md5 hash
df['md5'] = df['filename'].apply(lambda x: x.split('.')[0])

# load electronic artists df
df_electronic = pd.read_csv(reference_path)

# select only songs where best_version is true
df_electronic = df_electronic[df_electronic['best_version'] == True]

# select only songs where the source is electronic
df_electronic = df_electronic[df_electronic['source'] == 'good'] # 'good'

# merge the two data frames
df_electronic = pd.merge(df, df_electronic, on='md5')

# copy all files from the file_path column to the output folder
df_electronic['file_path'].apply(lambda x: shutil.copy(x, output_path))
```
